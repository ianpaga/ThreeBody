## Ballistic Model (3-body)

> Sepinsky+2010

- An **inertial** frame of reference $OXYZ$

  - Origin $O$ at the initial center ofmass of the binary prior to anymass overflow,
  - The $XY$-plane coinciding with the orbital plane
  - Positive $Z$-axis is parallel to the direction of the orbital angular momentum vector at the onset of mass overflow

### Basic Assumptions

- Circular orbit (**Kremer+2015**)

- Orbital angular velocity $\vec{\Omega}_P$

$$
  \left|\vec{\Omega}_{P}\right|\equiv\frac{(1+e)^{1 / 2}}{(1-e)^{3 / 2}} \frac{2 \pi}{P_{\mathrm{orb}}}=\frac{2 \pi}{P_{\mathrm{orb}}}
  $$
  
- Spin angular velocity

$$
\vec{\Omega}_{1}=f_{1} \vec{\Omega}_{P}, \quad \vec{\Omega}_{2}=f_{2} \vec{\Omega}_{P}
$$

- The surface of the star will conform to the equipotential surface which encloses the same volume as the spherical radius of the star. one of the stars is assumed to fill its effective Roche lobe due to the expansion of the star and/or the contraction of the orbit

- Donor - Star 1; Accretor - Star 2; Ballistic particle - 3

- Mass of the particle

$$
M_{3}=\int_{-P_{\mathrm{ob}} / 2}^{P_{\mathrm{ob} / 2}}\left(-\dot{M}_{1}\right) d t=-\frac{(1-e)^{3 / 2}}{(1+e)^{1 / 2}} \frac{P_{\mathrm{orb}}}{2 \pi} \dot{M}_{0}=\frac{P_{\mathrm{orb}}}{2 \pi} \dot{M}_{0}
$$

where $\dot{M}_{0}$ is the mass overflow rate

### The 3-body Problem
  $$
  \frac{d^{2} \vec{R}_{k}}{d t^{2}}=-\sum_{j=1 \atop j \neq k}^{3} \frac{G M_{j}}{\left|\vec{r}_{k j}\right|^{2}} \frac{\vec{r}_{k j}}{\left|\vec{r}_{k j}\right|} \quad(k=1,2,3)
  $$

  - For numerical convenience, we define several dimensionless quantities
  
  $$
  \begin{aligned} \vec{R}_{j}^{*} &=\vec{R}_{j} / r_{P, i}, \quad j=1,2,3 \\ M_{j}^{*} &=M_{j} / M, \quad j=1,2,3 \\ \vec{V}_{j}^{*} &=\vec{V}_{j} \sqrt{\frac{r_{P, i}}{G M}}, \quad j=1,2,3 \\ t^{*} &=t \sqrt{\frac{G M}{r_{P, i}^{3}}} \\ \vec{\Omega}_{P}^{*} &=\vec{\Omega}_{P} \sqrt{\frac{r_{P, i}^{3}}{G M}}\\
  \vec{r}_{L_{1}}^{*} &=\frac{\vec{r}_{L_{1}}}{r_{P, i}}\end{aligned}
  $$
  
    where $r_{P,i}\equiv a(1-e)=a$ is the initial (periastron) distance between the binary components

### Initial Condition

- Initial velocity of the ejected particle (spin + orbital velocity + thermal velocity)

  $$
  \vec{V}_{3}=\vec{\Omega}_{1} \times \vec{r}_{L_{1}}+\vec{V}_{1}+\vec{V}_{ej}
  $$
  
  where $\vec{r}_{L_{1}}$ is the vector from the center of mass of star 1 to the L1 point

- For numerical convenience,
    $$
    \begin{array}{l}{\vec{R}_{1, i}^{*} \equiv\left(X_{1, i}^{*}, Y_{1, i}^{*}, Z_{1, i}^{*}\right)=\left(\frac{1}{1+q_{i}}, 0,0\right)} \\ {\vec{R}_{2, i}^{*} \equiv\left(X_{2, i}^{*}, Y_{2, i}^{*}, Z_{2, i}^{*}\right)=\left(-\frac{q_{i}}{1+q_{i}}, 0,0\right)} \\  {\vec{R}_{3, i}^{*}=\vec{r}_{L_{1}, i}^{*}} \\ {\vec{V}_{1, i}^{*}= \vec{\Omega}_{P, i}^{*}\times \vec{R}_{1, i}^{*}} \\ {\vec{V}_{2, i}^{*}=\vec{\Omega}_{P, i}^{*}\times \vec{R}_{2, i}^{*}} \\ {\vec{V}_{3, i}^{*}=f_{1, i} \vec{\Omega}_{P, i}^{*} \times \vec{r}_{L_{1}, i}^{*}+ \vec{\Omega}_{P, i}^{*}\times \vec{R}_{1, i}^{*}}\end{array}\Rightarrow\begin{array}{l}{\vec{R}_{1, i}^{*} \equiv\left(X_{1, i}^{*}, Y_{1, i}^{*}\right)=\left(\frac{1}{1+q_{i}}, 0\right)} \\ {\vec{R}_{2, i}^{*} \equiv\left(X_{2, i}^{*}, Y_{2, i}^{*}\right)=\left(-\frac{q_{i}}{1+q_{i}}, 0\right)} \\  {\vec{R}_{3, i}^{*}=\vec{r}_{L_{1}, i}^{*}} \\ {\vec{V}_{1, i}^{*}=\left(0, \frac{1}{1+q_{i}}\Omega_{P,i}^*\right)} \\ {\vec{V}_{2, i}^{*}=\left(0, -\frac{q_{i}}{1+q_{i}}\Omega_{P,i}^*\right)} \\ {\vec{V}_{3, i}^{*}=\left(0,-f_{1,i}{r}_{L_{1}, i}^{*}\Omega_{P,i}^*+\frac{1}{1+q_{i}}\Omega_{P,i}^*\right)}\end{array}
    $$
    the dynamical equation
    $$
    \frac{d^{2} \vec{R}_{k}^{*}}{d\left(t^{*}\right)^{2}}=-\sum_{j=1 \atop j \neq k}^{3} \frac{M_{j}^{*}}{\left|\vec{r}_{k j}^{*}\right|^{3}} \vec{r}_{k j}^{*}\quad(k=1,2,3)
    $$

### Numerical method 

- Eighth-order Rungeâ€“Kutta ordinary differential equation solver

- Eccentricity & semi-major axis

$$
\begin{array}{l}{e^{2}=1-\frac{\left|\vec{r}_{\mathrm{rel}} \times \vec{v}_{\mathrm{rel}}\right|^{2}}{G\left(M_{1}+M_{2}\right) a}} \\ {\frac{1}{a}=\frac{2}{\left|\vec{r}_{\mathrm{rel}}\right|}-\frac{v_{\mathrm{rel}}^{2}}{G\left(M_{1}+M_{2}\right)}}\end{array}
$$
